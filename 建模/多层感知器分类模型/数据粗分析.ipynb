## 数据导入


```python
data_path='data.xlsx'
data=pd.read_excel(io=data_path,encoding='gbk')
def transform(row):
    if row['是否续保']=='是':
        row['是否续保']='1'
    else:
        row['是否续保']='0'
    return row
data_ture=data.apply(transform,axis=1)
data_ture['是否续保'] = data_ture['是否续保'].astype(np.float64)
```
